# Graph comparing different Simulators

## 1. Import libraries to make graph

```python
import pandas as pd
import matplotlib.pyplot as plt

import itertools
```

## 2. Graph data

1. Time execution and Memory usage in baseline circuit between different simulators

```python
df_mem = pd.read_csv("Record Data/Memory Usage/baseline.csv")
df_time = pd.read_csv("Record Data/Time Execution/baseline.csv")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

line_styles = ['-', '--' ]
markers = ['o', 's', 'D' ]

style_cycle_mem = itertools.cycle(line_styles)
marker_cycle_mem = itertools.cycle(markers)

for platform in df_mem['platform'].unique():
    subset = df_mem[df_mem['platform'] == platform]
    axes[0].plot(
        subset['number_of_qubits'],
        subset['memory_usage'],
        linestyle=next(style_cycle_mem),
        marker=next(marker_cycle_mem),
        label=platform
    )

axes[0].set_xlabel("Number of Qubits")
axes[0].set_ylabel("Memory Usage (MB)")
axes[0].set_title("Memory Usage vs Number of Qubits\n(baseline circuit)")
axes[0].legend()
axes[0].grid(True)

style_cycle_time = itertools.cycle(line_styles)
marker_cycle_time = itertools.cycle(markers)

for platform in df_time['platform'].unique():
    subset = df_time[df_time['platform'] == platform]
    axes[1].plot(
        subset['number_of_qubits'],
        subset['time_execution'],
        linestyle=next(style_cycle_time),
        marker=next(marker_cycle_time),
        label=platform
    )

axes[1].set_xlabel("Number of Qubits")
axes[1].set_ylabel("Time Execution (s)")
axes[1].set_title("Time Execution vs Number of Qubits\n(baseline circuit)")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
